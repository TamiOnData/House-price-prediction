```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error, r2_score 
```

```python
#load the cleaned dataset
data = pd.read_csv("cleaned_housing.csv")
```

```python
#explore the dataset
data.head()
```

```python
#split into features{X) and target(y)
X = data.drop('median_house_value', axis=1)
#features. so we pick all columns except house_value
y = data['median_house_value']
```

```python
#train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.2, #20% of dataset should be for testing. 80/20 split
    random_state = 89 #seeding value
) 
```

```python
X_train
```

```python
X_test
```

```python
#create and train the model
model = LinearRegression()
model.fit(X_train, y_train) 
```

```python
#make predictions
predictions = model.predict(X_test)
```

```python
predictions
```

```python
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
```

```python
r2
mse
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
ridge = Ridge(alpha=1.0)#create model
ridge.fit(X_train_scaled, y_train)#train it
```

```python
y_pred_ridge = ridge.predict(X_test_scaled)
```

```python
#compare prediction
print("Actual: ", y_test.values)
print("Predicted: ", y_pred_ridge)
```

```python
#evaluate prediction
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
mse_ridge
```
